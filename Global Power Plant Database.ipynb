```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error

```

```python
# Load the dataset
data_url = "https://github.com/wri/global-power-plant-database/raw/master/source_databases_csv/database_IND.csv"
df = pd.read_csv(data_url)
```

```python
df
```

```python
# Data preprocessing
# For simplicity, let's drop rows with missing values for this example
df.dropna(inplace=True)
```

```python
# Select features and target for fuel type prediction
X_fuel = df[['capacity_mw', 'latitude', 'longitude']]
y_fuel = df['primary_fuel']
```

```python
# Split the data into training and testing sets
X_train_fuel, X_test_fuel, y_train_fuel, y_test_fuel = train_test_split(X_fuel, y_fuel, test_size=0.2, random_state=42)
```

```python
# Train a Random Forest Classifier for fuel type prediction
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_fuel, y_train_fuel)
```

```python
# Predict fuel types
y_pred_fuel = rf_classifier.predict(X_test_fuel)
```

```python
# Evaluate the classifier
accuracy = accuracy_score(y_test_fuel, y_pred_fuel)
print("Accuracy for fuel type prediction:", accuracy)
```

```python
# Select features and target for capacity prediction
X_capacity = df[['latitude', 'longitude', 'primary_fuel']]
y_capacity = df['capacity_mw']
```

```python
# Encode categorical variable for capacity prediction
X_capacity = pd.get_dummies(X_capacity)
```

```python
# Split the data into training and testing sets
X_train_capacity, X_test_capacity, y_train_capacity, y_test_capacity = train_test_split(X_capacity, y_capacity, test_size=0.2, random_state=42)
```

```python
# Train a Random Forest Regressor for capacity prediction
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train_capacity, y_train_capacity)
```

```python
# Predict capacity
y_pred_capacity = rf_regressor.predict(X_test_capacity)
```

```python
# Evaluate the regressor
mse = mean_squared_error(y_test_capacity, y_pred_capacity)
print("Mean Squared Error for capacity prediction:", mse)
```
